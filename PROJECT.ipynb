```python
#DETERMINE THE DISTRIBUTION OF EMPLOYEES ACROSS EACH TEAM
team_distribution=data["Team"].value_counts().sort_values(ascending=False)
#value_counts()->count the number of occurrences of each unique value in a Series (column) of a DataFrame.
#sort_values(ascending=False)->sort values in descending order
print(team_distribution)
```

```python
#CALCULATE THE PERCENTAGE SPLIT RELATIVE TO THE TOTAL NUMBER OF EMPLOYEES
total_employees = len(data)#len()-returns the number of items in an object.object can be a string, list, tuple, dictionary, set, or any collection.
print(total_employees)
```

```python
# CALCULATE THE PERCENTAGE SPLIT FOR EACH TEAM
team_percentage = (team_distribution / total_employees) * 100
#print(team_percentage )
print(team_percentage.round(2) )# round a number to the nearest integer or to a specified number of decimal places.
```

```python
# Create a bar chart
plt.figure(figsize=(10,6))
plt.bar(team_percentage.index,team_percentage.values)
plt.xlabel('Team')
plt.ylabel('Percentage of Employees')
plt.title('Distribution of Employees by Team')
plt.show()
```
